```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
df = pd.read_csv('OnlineNewsPopularity.csv')

# remove left space from column headers
df.columns = df.columns.str.replace(' ', '')

#drop non-predictive items
df=df.drop(['url', 'timedelta'], axis=1)

#drop negative values for Naive Bayes
df=df.drop(['kw_min_min', 
           'kw_avg_min',
           'kw_min_avg',
           'global_sentiment_polarity',
           'avg_negative_polarity',
           'min_negative_polarity',
           'max_negative_polarity',
           'title_sentiment_polarity',], axis=1)


# replace shares count with dummy class
df.loc[df.shares >= 10000, 'Viral'] = 1
df.loc[df.shares < 10000, 'Viral'] = 0 
df=df.drop(['shares'], axis=1)

df.head()
```

```python
X = []
y= []
X = df.iloc[:, :49]
y = df.iloc[:, 50]

# Split dataset in training and test datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))
```

```python
target_names = ["not viral", "viral"]

# Calculate classification report
from sklearn.metrics import classification_report
predictions = clf.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
