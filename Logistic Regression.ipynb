```python
import numpy as np
import pandas as pd
import csv
```

```python
# Importing the dataset
df = pd.read_csv('OnlineNewsPopularity.csv')

# remove left space from column headers
df.columns = df.columns.str.replace(' ', '')

# drop non-predictive items
df=df.drop(['url', 'timedelta'], axis=1)

# replace shares count with "Viral" class
df.loc[df.shares >= 10000, 'Viral'] = 'True' 
df.loc[df.shares < 10000, 'Viral'] = 'False' 
df=df.drop(['shares'], axis=1)

df.head()
```

```python
#separate variables into X (independents) and Y (dependent)
X = []
y= []
X = df.iloc[:, :57]
y = df.iloc[:, 58]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(tol=0.1, solver='saga') #'tol' and 'sag' optimizes for multivariate
```

```python
model.fit(X_train, y_train)
```

```python

print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
target_names = ["not viral", "viral"]

# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
