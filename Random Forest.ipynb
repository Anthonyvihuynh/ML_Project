```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv

from sklearn.ensemble import RandomForestClassifier
```

```python
# Importing the dataset
df = pd.read_csv('OnlineNewsPopularity.csv')

# remove left space from column headers
df.columns = df.columns.str.replace(' ', '')

# drop non-predictive items
df=df.drop(['url', 'timedelta'], axis=1)

# replace shares count with "Viral" class
df.loc[df.shares >= 11500, 'Viral'] = 'True' 
df.loc[df.shares < 11500, 'Viral'] = 'False' 
df=df.drop(['shares'], axis=1)

df.head()
```

```python
#separate descriptive variables X (independents) and Y (dependent)
target = df["Viral"]
data = df.drop("Viral", axis=1)
feature_names = data.columns
data.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
target_names = ["not viral", "viral"]

# Calculate classification report
from sklearn.metrics import classification_report
predictions = rf.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
